```python
import pandas as pd
import numpy as np
import os
import plotly.express as px
import seaborn as sns
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
manufacturer_df = pd.read_csv("../Resources/manufacturer_breakdown.csv")
manufacturer_df = manufacturer_df[['location','date','CanSino','Johnson&Johnson','Moderna','Oxford/AstraZeneca','Pfizer/BioNTech','Sinopharm/Beijing','Sinovac','Sputnik V']]
```

```python
manufacturer_df["location"].unique()
```

```python
manufacturer_df.fillna(float(0))
```

```python
#Make eight mini-line graphs to show vaccines by vendor
by_vendor_line = make_subplots(rows=4,cols=2)
by_vendor_line =px.line(manufacturer_df, x='date', y= 'Johnson&Johnson', color ='location', title = 'Johnson & Johnson breakdown usage by country')
by_vendor_line.show()
```

```python
by_vendor_line = make_subplots(rows=4,cols=2)
by_vendor_line =px.line(manufacturer_df, x='date', y= 'Pfizer/BioNTech', color ='location', title = 'Pfizer/BioNTech breakdown usage by country')
by_vendor_line.show()
```

```python
by_vendor_line = make_subplots(rows=4,cols=2)
by_vendor_line =px.line(manufacturer_df, x='date', y= 'Moderna', color ='location', title = 'Moderna breakdown usage by country')
by_vendor_line.show()
```

```python
last_dates_df = pd.DataFrame()
for country in manufacturer_df["location"].unique():
    last_dates_df= last_dates_df.append(manufacturer_df[manufacturer_df["location"] == country].sort_values("date").iloc[-1:,:])

last_dates_df
```

```python
import matplotlib.pyplot as plt
loc = last_dates_df.groupby('location')['CanSino', 'Moderna','Oxford/AstraZeneca', 'Pfizer/BioNTech','Sinopharm/Beijing','Sinovac','Sputnik V'].max()
loc.plot(kind= 'bar', figsize = (40,20), fontsize = 25)
```
